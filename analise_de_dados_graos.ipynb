```python
import numpy as np
import pandas as pd
```

Carregando o conjunto de dados em um quadro de dados e traduzindo as colunas usando Pandas

```python
graos=pd.read_csv('/content/agricultural_raw_material (1).csv')
graos.columns = ["Mês", "Preço de lã grossa", "Variação % do preço da lã grossa", "Preço Copra", "Alteração percentual do preço da Copra", "Preço do algodão", "Alteração percentual do preço do algodão", "Preço de lã fina", "Variação % do preço da lã fina", "Preço de registro rígido", "Alteração percentual de preço de log rígido", "Preço de madeira serrada dura", "Variação % do preço da madeira serrada", "Ocultar preço", "Ocultar variação percentual de preço", "Preço da madeira compensada", "Variação % do preço do compensado", "Preço da Borracha", "Alteração percentual do preço da borracha", "Preço do Softlog", "Alteração percentual de preço do Softlog", "Preço de madeira serrada macia", "Variação % do preço da madeira serrada macia", "Preço da polpa de madeira", "Preço da polpa de madeira % Chang"]

#Verificando valores nulos de cada coluna
graos.info
graos.isnull().sum()

```

Lidando com dados ausentes, incorretos e inválidos

```python
graos= graos.replace('%', '', regex = True)
graos= graos.replace(',' , '' ,regex = True)
graos= graos.replace('-', '', regex = True)
graos= graos.replace('', np.nan)
graos= graos.replace('MAY90', np.nan)

# Eliminando linhas com valores NaN

graos.info
graos.isnull().sum()
```

```python
# Convertendo tipo de dados para float
lst = price_list = [
   "Preço de lã grossa", "Variação % do preço da lã grossa", "Preço Copra", "Alteração percentual do preço da Copra", "Preço do algodão", "Alteração percentual do preço do algodão", "Preço de lã fina", "Variação % do preço da lã fina", "Preço de registro rígido", "Alteração percentual de preço de log rígido", "Preço de madeira serrada dura", "Variação % do preço da madeira serrada", "Ocultar preço", "Ocultar variação percentual de preço", "Preço da madeira compensada", "Variação % do preço do compensado", "Preço da Borracha", "Alteração percentual do preço da borracha", "Preço do Softlog", "Alteração percentual de preço do Softlog", "Preço de madeira serrada macia", "Variação % do preço da madeira serrada macia", "Preço da polpa de madeira", "Preço da polpa de madeira % Chang"
]
graos[lst] = graos[lst].astype("float")
graos.dtypes
```

```python
graos.head()
```

**Colunas de data e horacolunas**

formatando a coluna datetime e definindo-a como índice para o conjunto de dados

```python
graos['Mês'] = pd.to_datetime(graos['Mês'].str.upper(), format='%b%y', yearfirst=False)

# Indexando pelo mês
graos = graos.set_index('Mês')
```

Importando **matplotlib.pyplot e seaborn.**








```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

**Mapa de calor**

---

Um mapa de calor é uma técnica de visualização de dados que mostra a magnitude de um fenômeno por meio de cor em duas dimensões.

```python
raw_data=['Preço de lã grossa', 'Preço Copra', 'Preço do algodão', 'Preço de lã fina',
    'Preço de registro rígido', 'Preço de madeira serrada dura',
    'Preço da madeira compensada', 'Preço da Borracha', 'Preço do Softlog',
    'Preço de madeira serrada macia', 'Preço da polpa de madeira'

]

corrmat = graos[raw_data].corr()


fig = plt.figure(figsize = (12, 9))


mask = np.triu(np.ones_like(corrmat, dtype=bool))
sns.heatmap(corrmat, vmax = .8,mask=mask, square = True, annot = True, annot_kws={"size": 7})
plt.show()
```

# **Análise:**

  O mapa de calor mostra a correlação entre as matérias-primas maior o valor correlacionado maior chance de serem duas matérias-primas relacionadas, mas não necessariamente

---



*   **Madeira serrada:**
Madeira serrada macia: preços baixos, com destaque para o preço do Softlog (verde escuro).
*  **Madeira serrada dura:** preços mais altos, com destaque para o preço do registro rígido (vermelho escuro).

*   **Madeira compensada:** preços relativamente baixos, com algumas variações (verde claro a amarelo).
*   **Borracha:** preços altos, com destaque para o preço da Borracha (vermelho escuro).

**Outros produtos:**

1.   Copra: preço baixo, com ligeira queda recente (verde).
Preço de la grossa e fina: preços relativamente altos, estáveis (vermelho).

A partir deste gráfico podemos dizer que quase não há relação entre a variação percentual dos preços das matérias-primas

* o valor negativo implica que duas variáveis ​​estão negativamente correlacionadas (um aumento, outro decréscimo)
* Zero implica nenhuma relação
* caso contrário, maior o valor maior a chance de relação.

# **Variação normal do preço de cada matéria-prima**

```python
changelist=['Variação % do preço da lã grossa', 'Alteração percentual do preço da Copra',
    'Alteração percentual do preço do algodão', 'Variação % do preço da lã fina',
    'Alteração percentual de preço de log rígido', 'Variação % do preço da madeira serrada',
    'Ocultar variação percentual de preço', 'Variação % do preço do compensado',
    'Alteração percentual do preço da borracha', 'Alteração percentual de preço do Softlog',
    'Variação % do preço da madeira serrada macia', 'Preço da polpa de madeira % Chang']
for i in range(len(changelist)):
    plt.figure(figsize=(12,12))
    graos[changelist[i]].hist(figsize=(11, 9), linewidth=1)
    plt.xlabel('% Change')
    plt.ylabel('count')
    plt.legend(changelist[i:],loc='upper center',bbox_to_anchor=(1.2,1))
```

# **Matéria-prima que tem o menor preço ao longo dos anos**
